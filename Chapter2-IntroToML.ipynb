```python
#Let's fill in the missing values with the median
median_age = df_titanic_features['Age'].median()
print(median_age)
```

```python
df_titanic_features['Age'].fillna(median_age, inplace=True)
```

What Pandas is trying to teach me: Avoid inplace=True on chained objects. Other words, modifying it the way I did initally will create a copy of the dataset. This will fail in pandas 3.0 and I will be working with the incorrect dataset unless I specify I am targeting the copy of which I created. 

```python
df_titanic_features.fillna({'Age':median_age}, inplace=True)
```

```python
#fill the missing values of the Embarked attribute with the most common value in the column
embarked_value_counts = df_titanic_features['Embarked'].value_counts(dropna=True)
print(embarked_value_counts)
most_common_value = embarked_value_counts.index[0]
print(most_common_value)
```

```python
df_titanic_features.fillna({'Embarked':most_common_value}, inplace=True)
```

```python
#create a boolean feature "CabinIsKnown"
#This feature will have True if its in the Cabin column 
#Does not have missing data
df_titanic_features['CabinIsKnown'] = ~df_titanic_features.Cabin.isnull()

#Drop Cabin from the dataset
df_titanic_features.drop(['Cabin'], axis=1, inplace=True)
```

```python
#Let's check to see if the columns were updated correctly
df_titanic_features.head()
```

```python
#Display the number of missing values 
df_titanic_features.isnull().sum()
```

Alright, we have succesffully updated all of the missing values from the columns. 

```python
"""
Let's combine the SibSp column and the Parch column so that we can get a better understadning of the entire family size. 
We'll combine these two columns through addition. 
"""
df_titanic_features['FamilySize'] = df_titanic_features['SibSp'] + df_titanic_features['Parch']
```

```python
df_titanic_features.head()
```

Any changes made to the training data should also be carried out on the TEST DATA!!!

```python
#Generate New Categorical Feature
bins_age = [0,20,30,40,50,150]
labels_age = ['<20','20-30','30-40','40-50','>50']
```

```python
df_titanic_features['Age_Category'] = pd.cut(df_titanic_features.Age,
                                             bins=bins_age,
                                             labels=labels_age,
                                             include_lowest=True)
```

```python
df_titanic_features.head(10)
```

```python
df_titanic_features['FareCategory'] = pd.qcut(df_titanic_features.Fare,
                                              q=4,
                                              labels=['Q1','Q2','Q3','Q4'])
```

```python
df_titanic_features.head()
```

```python
df_titanic_target.head()
```

```python
#Generate new feature NormalizedAge using MinMaxScaler
from sklearn import preprocessing

minmax_scaler = preprocessing.MinMaxScaler()
ndNormalizedAge = minmax_scaler.fit_transform(df_titanic_features[['Age']].values) #Use double brackets because fit_transforms requires a 2D array
df_titanic_features['NormalizedAge'] = pd.DataFrame(ndNormalizedAge)

#generate new features StandardizedAge using StandardScaler
standard_scaler = preprocessing.StandardScaler()
ndStandarizedAge = standard_scaler.fit_transform(df_titanic_features[['Age']].values)
df_titanic_features['StandardizedAge'] = pd.DataFrame(ndStandarizedAge)

#Now let's create a histogram of the Normalized and Standardized dataframes we created.
df_titanic_features[['Age','NormalizedAge','StandardizedAge']].hist(figsize=(10,10), bins=5)
```

One-Hot Encoding Categorical Features - Converting categorical features into numeric features.

```python
#use one-hot encoding to convert categorical attributes into binary numeric attributes
df_titanic_features = pd.get_dummies(df_titanic_features, columns=['Sex','Embarked', 'CabinIsKnown','Age_Category','FareCategory'])

#Display the columns of the dataframe. 
print(df_titanic_features.columns.values)
```
