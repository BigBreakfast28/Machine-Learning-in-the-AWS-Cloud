```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#Figure1
%matplotlib inline
x = np.linspace(1,7,10)

y1 = 4*x*x + 2*x +2
y2 = 3*x + 4

figure1,ax= plt.subplots(figsize=(5,5))

ax.plot(x,y1)
ax.plot(x,y2)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

plt.suptitle('Figure One')
figure1.set_facecolor('#FFFF8F')

#Figure2
a = np.linspace(2,18, num=6)

b1= 4*a*a + 8*a + 3
b2 = 9*a + 2

figure2, ax = plt.subplots(figsize=(5,5))

ax.plot(a,b1)
ax.plot(a,b2)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

figure2.suptitle('Figure Two')
figure2.set_facecolor('#FFDF00')
plt.show()
```

```python
e = np.linspace(1,20, num=40)
f1 = 3*e*e + 5*e + 3
f2 = 10*e + 20

figure3, ax = plt.subplots(figsize=(7,7))
ax.grid(True)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')

ax.plot(e,f1)
ax.plot(e,f2)

figure3.suptitle('Figure Three')
figure3.set_facecolor('#CBC3E3')
```
